```python
nitrogen=int(input("enter the soil nitrogen value in mg\kg:"))


if nitrogen<100:
  print("application of high nitrogen fertilizer\ncost:rs.2000\kg")
  print("message: frequent apply of nitrogen fertilizer")
elif  100<=nitrogen<=250:
  print("apply medium nitrogen fertilizer\ncost:rs.1000\kg")
else:
  print("no fertilizer needed,soil is healthy")
```

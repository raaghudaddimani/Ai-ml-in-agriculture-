```python
marks=[56,78,90,45,88]
max_marks=max(marks)
total=0
for m in marks:
  total=total+m
  average=total/len(marks)
  count=0
  for m in marks:
    if m>=60:
      count=count+1
print("max:",max_marks)
print("average:",average)
print("count>=60:",count)


```

```python
def average_marks(marks_list):
   total=0
   for m in marks_list:
     total=total+m
     average=total/len(marks_list)
     return average
data=[20,30,40]
print("average=",average_marks(data))
```
